```python
with open("./day_06_input.txt") as f:
    lines = [[x for x in i.replace("\n", "")] for i in f.readlines()]
```

```python
lines
```

```python
cur_x = 0
cur_y = 0
cur_symbol = "^"
position_found = False

while not position_found:
    for y_index, line in enumerate(lines):
        for x_index, pos in enumerate(line):
            if pos == "^":
                cur_x = x_index
                cur_y = y_index
                position_found = True

print(cur_y, cur_x)
```

```python
not_finished = True
visited_spots = 0
next_dir = {
    "^": ">",
    ">": "v",
    "v": "<",
    "<": "^"
}

additions = {
    "^": (-1, 0),
    ">": (0, 1),
    "v": (1, 0),
    "<": (0, -1)
}

while not_finished:
    try:
        next_x = cur_x + additions[cur_symbol][1]
        next_y = cur_y + additions[cur_symbol][0]
        if lines[next_y][next_x] == "#":
            cur_symbol = next_dir[cur_symbol]
            continue
        elif lines[next_y][next_x] == ".":
            lines[cur_y][cur_x] = "X"
            visited_spots += 1
            cur_x += additions[cur_symbol][1]
            cur_y += additions[cur_symbol][0]
            lines[cur_y][cur_x] = cur_symbol
        elif lines[next_y][next_x] == "X":
            lines[cur_y][cur_x] = "X"
            cur_x += additions[cur_symbol][1]
            cur_y += additions[cur_symbol][0]
            lines[cur_y][cur_x] = cur_symbol
    except IndexError:
        visited_spots += 1
        not_finished = False

print(visited_spots)
```
